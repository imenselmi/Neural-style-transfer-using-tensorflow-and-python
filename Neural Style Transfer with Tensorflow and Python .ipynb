```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model=hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_img(img_path):
  img=tf.io.read_file(img_path)
  img=tf.image.decode_image(img,channels=3)
  img=tf.image.convert_image_dtype(img,tf.float32)
  img=img[tf.newaxis,:]
  return img
```

```python
content_image=load_img('/content/drive/MyDrive/Colab Notebooks/MTC/labs/deep learning/Neural style transform/Python-Neural-Style-Transfer-main/264554063_2495693960575141_4481611338210417796_n.jpg')
style_image=load_img('/content/drive/MyDrive/Colab Notebooks/MTC/labs/deep learning/Neural style transform/Python-Neural-Style-Transfer-main/starrynight.jfif')
```

```python
plt.imshow(np.squeeze(style_image))
plt.show
```

```python
stylized_image=model(tf.constant(content_image),tf.constant(style_image))[0]
```
